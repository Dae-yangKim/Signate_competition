```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '/content/drive/MyDrive/Signate/2023.08.30 MUFG/data/'
```

```python
train_df = pd.read_csv(path + 'train.csv')
test_df = pd.read_csv(path + 'test.csv')
user_df = pd.read_csv(path + 'user.csv')
card_df = pd.read_csv(path + 'card.csv')
```

```python
# Merge

merge_train_df = pd.merge(train_df , card_df , on = ['user_id' , 'card_id'] , how = 'left')
merge_train_df = pd.merge(merge_train_df , user_df , on = 'user_id' , how = 'left')
merge_train_df.drop(['index'] , axis = 1 , inplace = True)
```

```python
merge_test_df = pd.merge(test_df , card_df , on = ['user_id' , 'card_id'] , how = 'left')
merge_test_df = pd.merge(merge_test_df , user_df , on = 'user_id' , how = 'left')
merge_test_df.drop(['index'] , axis = 1 , inplace = True)
```

```python
print(f'Data set size before merge is {train_df.shape}')
print(f'Data set size after merge is {merge_train_df.shape}')
```

```python
print(f'Test Data set size before merge is {test_df.shape}')
print(f'Test Data set size after merge is {merge_test_df.shape}')
```

```python
merge_train_df.to_csv(path + 'merge_train_df.csv')
merge_test_df.to_csv(path + 'merge_test_df.csv')
```
