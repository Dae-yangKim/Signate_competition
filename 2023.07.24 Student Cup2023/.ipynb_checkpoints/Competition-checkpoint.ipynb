```python
def change_columns(data , col):

    for i in range(8):

        data.rename(columns = {f'col_{i}' : f'{col}_{i}'} , inplace = True)
```

```python
def data_extraction(df , col , test = False):

    length = len(df)
    data = df.groupby([col]).size() / length * 100
    data = pd.DataFrame(data)
    data.rename(columns = {0 : 'percentage'} , inplace = True)
    data = data.sort_values(['percentage'] , ascending = False)
    under_data = data[data['percentage'] < 0.1]

    if not test:
        print(f'All Data : {data.shape}')
        print(f'Filtered Data : {under_data.shape}')

        for idx in range(len(df)):

            sub = df.loc[idx , col]

            if sub in under_data.index:

                df.loc[idx , col] = 'other'

    else:
        print(f"All Data (test) : {data.shape}")
        print(f"Filtered Data (test) : {under_data.shape}")

        for idx in range(len(df)):

            sub = df.loc[idx , col]

            if sub in under_data.index:

                df.loc[idx , col] = 'other'
```

```python
# transmission

transmission_map = {
    'other' : 1 ,
    'manual' : 2 ,
    'automatic' : 3
}

df['transmission'] = df['transmission'].map(transmission_map)
test_df['transmission'] = test_df['transmission'].map(transmission_map)
```

```python
from hyperopt import hp , fmin , tpe , Trials , STATUS_OK

# GBR
GBR_search_space = {
    'loss' : hp.choice('loss' , ['squared_error' , 'absolute_error']) ,
    'n_estimators' : hp.choice('n_estimators' , [100 , 200 , 300]) ,
    'learning_rate' : hp.choice('learning_rate' , [0.1 , 0.05 , 0.01]) ,
    'max_depth' : hp.choice('max_depth' , [4 , 8]) ,
    'min_samples_leaf' : hp.choice('min_samples_leaf' , [100 , 150]) ,
    'max_features' : hp.choice('max_features' , [0.3 , 0.1])
}

# XGBR
XGBR_search_space = {
    'n_estimators' : hp.choice('n_estimators' , [100 , 200 , 300]) ,
    'learning_rate' : hp.choice('learning_rate' , [0.1 , 0.05 , 0.01]) ,
    'max_depth' : hp.choice('max_depth' , [4 , 8]) ,
    'gamma' : hp.choice('gamma' , list(range(0 , 11))) ,
    'subsample' : hp.choice('subsample' , [0.5 , 0.6 , 0.7 , 0.8 , 0.9 , 1])
}

# LGBR
LGBR_search_space = {
    'n_estimators' : hp.choice('n_estimators' , [100 , 200 , 300]) ,
    'learning_rate' : hp.choice('learning_rate' , [0.1 , 0.05 , 0.01]) ,
    'max_depth' : hp.choice('max_depth' , [4 , 8]) ,
    'gamma' : hp.choice('gamma' , list(range(0 , 11))) ,
    'subsample' : hp.choice('subsample' , [0.5 , 0.6 , 0.7 , 0.8 , 0.9 , 1])
}

# KNN
KNN_search_space = {
    'weights' : hp.choice('weights' , ['uniform' , 'distance']) ,
    'n_neighbors' : hp.choice('n_neighbors' , list(range(2 , 9 , 3))) ,
    'algorithm' : hp.choice('algorithm' , ['auto' , 'ball_tree' , 'kd_tree' , 'brute'])
}
```

```python
def GBR_objective_func(search_space):

    model = GBR(
        loss = search_space['loss'] ,
        n_estimators = search_space['n_estimators'] ,
        learning_rate = search_space['learning_rate'] ,
        max_depth = search_space['max_depth'] ,
        min_samples_leaf = search_space['min_samples_leaf'] ,
        max_features = search_space['max_features']
    )

    score = cross_val_score(
        model , x_train , y_train ,
        scoring = 'neg_mean_absolute_percentage_error' ,
        cv = 5 , n_jobs = 4
    ).mean()

    return {'loss' : -score , 'status' : STATUS_OK}
```

```python
GBR_trials = Trials()

GBR_best = fmin(
    fn = GBR_objective_func ,
    space = GBR_search_space ,
    algo = tpe.suggest ,
    max_evals = 10 ,
    trials = GBR_trials
)

GBR_best
```

```python
def XGB_objective_func(search_space):

    model = XGBR(
        n_estimators = search_space['n_estimators'] ,
        learning_rate = search_space['learning_rate'] ,
        max_depth = search_space['max_depth'] ,
        gamma = search_space['gamma'] ,
        subsample = search_space['subsample']
    )

    score = cross_val_score(
        model , x_train , y_train ,
        scoring = 'neg_mean_absolute_percentage_error' ,
        cv = 5 , n_jobs = 4
    ).mean()

    return {'loss' : -score , 'status' : STATUS_OK}
```

```python
XGB_trials = Trials()

XGB_best = fmin(
    fn = XGB_objective_func ,
    space = XGBR_search_space ,
    algo = tpe.suggest ,
    max_evals = 10 ,
    trials = XGB_trials
)

XGB_best
```
